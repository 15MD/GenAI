```python
import fitz  # PyMuPDF

def read_pdf(file_path):
    doc = fitz.open(file_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

# Example usage
# Provide the relative file path
file_path = "C:\\Users\\10694387\\OneDrive - LTIMindtree\\Documents\\Pytorch\\GenAI\\Data\\LTI_MohitDeshmukh_DS (1).docx"
# file_path = "LTIMindtree\\Documents\\Pytorch\\GenAI\\Data\\LTI_MohitDeshmukh_DS (1).docx"

resume_content = read_pdf(file_path)

print(resume_content)
```

```python
from docx import Document

def read_docx(file_path):
    doc = Document(file_path)
    text = ""
    for paragraph in doc.paragraphs:
        text += paragraph.text + "\n"
    return text

# Example usage
resume_content = read_docx(file_path)
print(resume_content)
```

```python
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_core.documents import Document
import json

# Initialize embeddings
embeddings = HuggingFaceEmbeddings(model="sentence-transformers/all-MiniLM-L6-v2")

# Initialize ChromaDB
vector_store = Chroma(
    collection_name="job_applications",
    embedding_function=embeddings,
    persist_directory="./chroma_langchain_db"
)

# Read resume from .docx
resume_content = read_docx(file_path)

# Add your resume to ChromaDB
resume_document = Document(page_content=resume_content, metadata={"source": "resume"})
vector_store.add_documents([resume_document])

# Define the prompt template
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain_community.llms import Ollama

prompt_template = PromptTemplate(
    input_variables=["job_description", "resume"],
    template="""
    Based on the following job description and resume, generate a professional email template to apply for the job.

    Job Description:
    {job_description}

    Resume:
    {resume}

    Email Template:
    """
)

# Initialize the LLM
# llm = OpenAI(model="text-davinci-003")
llm=Ollama(model="llama2")

# Create the chain
chain = LLMChain(llm=llm, prompt=prompt_template)

# Fetch job descriptions from the saved JSON file
with open('jobs.json', 'r') as f:
    jobs = json.load(f)

# Generate email templates for each job
for job in jobs:
    job_description = job['description']
    email_template = chain.run(job_description=job_description, resume=resume_content)
    print(email_template)
```
